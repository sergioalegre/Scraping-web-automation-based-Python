```python
#Basado: https://www.udemy.com/course/big-data-data-science-python-web-scraping-2019-achirou-alvaro-chirou/learn/lecture/18156884#notes

import urllib
from bs4 import BeautifulSoup

direccion = 'https://www.lmneuquen.com/se-hizo-pasar-dueno-un-terreno-y-lo-vendio-internet-lo-acusaron-estafa-n685827'
codigo = urllib.request.urlopen(direccion)
var1 = BeautifulSoup(codigo)
p = var1('p') 

texto = '' 
for x in p:
    if (len(x.attrs)) == 0: 
        a= x.contents[0] 
        texto = str(texto) +  str(a)

almacenar_contar = dict() 
texto = texto.replace(',','').replace('.','').replace(':','').replace('?','').replace('(','').replace(')','') #quitamos los signos para quedarnos solo con las palabras
letras = texto.lower().split() #todo a minusculas y corto palabra a palabra (con split)
for x in letras:
    if len(x)>3: #me quedo solo con palabras de mas de 3 letras
        almacenar_contar[x] = almacenar_contar.get(x,0) + 1 
        
import pandas as pd
pd.DataFrame(list(almacenar_contar.items()), columns=['palabra', 'contador']).sort_values('contador', ascending=False )
```
