```python
#queremos extraer solo los enlaces (href) de una web
import urllib
from bs4 import BeautifulSoup

pagina_web = input('Enter - ') #para que pida al usuario que página analizar
codigo = urllib.request.urlopen(pagina_web)
var1 = BeautifulSoup(codigo)

etiqueta = var1('a') #a de <a href= ...
for x in etiqueta:
    print(x.get('href')) #nos quedamos solo con la parte de href
```

```python
#lo mismo con mas detalle
import urllib
from bs4 import BeautifulSoup

pagina_web = input('Enter - ')
codigo = urllib.request.urlopen(pagina_web)
var2 = BeautifulSoup(codigo)

etiqueta = var2('a')
for x in etiqueta:
    print('Etiqueta "a" completa:', x)
    print('Dirección:', x.get('href'))
    print('Contenido:', x.contents)
    print('Atributo:', x.attrs)
    print('\n')
```
