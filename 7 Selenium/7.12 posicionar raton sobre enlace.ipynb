```python
#Basado https://www.udemy.com/course/big-data-data-science-python-web-scraping-2019-achirou-alvaro-chirou/learn/lecture/21322492#notes

#en eswte ejemplo vamos a posicionarnos sobre un enlace como si pusieramos el raton sobre el

from selenium.webdriver import ActionChains #require de esta libreria especial
from selenium import webdriver
import time

chrome_driver = webdriver.Chrome("chromedriver.exe")
enlace = chrome_driver
enlace.get("https://educa.mastech.academy/")
time.sleep(2)

hipervinculo = enlace.find_element_by_link_text("Packs Cursos")

posicionar = ActionChains(enlace).move_to_element(hipervinculo)
posicionar.perform()
```
